```python
# Cột store_region
def store_region(df, col_name):
    from pyspark.sql import SparkSession
    from pyspark.sql.functions import initcap

    spark = SparkSession.builder.appName("CapitalizeExample").getOrCreate()
   
    return df.withColumn(col_name, initcap(df[col_name]))

# cột store_rating
def store_rating(df, col_name="store_rating"):
    from pyspark.sql.functions import regexp_replace, col, lower
    from pyspark.sql.types import FloatType
    # B1: chuẩn hóa dấu phẩy thành dấu chấm
    df = df.withColumn(col_name, regexp_replace(col(col_name), ",", "."))
    
    # B2: đưa về chữ thường
    df = df.withColumn(col_name, lower(col(col_name)))
    
    # B3: mapping chữ -> số
    mapping = {
        "one": "1",
        "two": "2",
        "three": "3",
        "four": "4",
        "five": "5"
    }
    for word, num in mapping.items():
        df = df.withColumn(
            col_name,
            regexp_replace(col(col_name), word, num)
        )
    
    # B4: ép kiểu float
    df = df.withColumn(col_name, col(col_name).cast(FloatType()))
    
    return df

#cột product_sku
def product_sku(df, col_name="product_sku"):
    from pyspark.sql.functions import col, trim, regexp_replace
    # B1: loại bỏ khoảng trắng ở đầu và cuối
    df = df.withColumn(col_name, trim(col(col_name)))
    
    # B2: loại bỏ khoảng trắng thừa bên trong (nếu có nhiều space giữa các ký tự)
    df = df.withColumn(col_name, regexp_replace(col(col_name), "\s+", ""))
    
    return df

# cột product_name
def clean_product_name(df, col_name="product_name"):
    from pyspark.sql.functions import trim, regexp_replace, col, initcap
    # B1: loại bỏ khoảng trắng thừa ở đầu/cuối
    df = df.withColumn(col_name, trim(col(col_name)))
    
    # B2: thay nhiều khoảng trắng bằng 1 khoảng trắng
    df = df.withColumn(col_name, regexp_replace(col(col_name), "\s+", " "))
    
    # B3: chuẩn hóa viết hoa chữ cái đầu (Product 102, Product 300...)
    df = df.withColumn(col_name, initcap(col(col_name)))
    
    return df



```
